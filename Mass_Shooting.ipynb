```python
total_victims_state=dataset.groupby('State').sum()
print('Total Victims=', total_victims_state['#Total'].sum())
total_victims_state.sort_values(by='#Total', ascending= False)
```

```python
total_victims_state.sort_values(by='#Total', ascending= False).head()
```

```python
total_victims_state.sort_values(by='#Total', ascending= False).tail()
```

```python
plt.subplots(figsize=(15,6))
cr=total_victims_state['#Total'].sort_values(ascending= False)
ax=cr.plot.bar()
ax.set_xlabel('State')
ax.set_ylabel('Total Number = Dead + Injured')
ax.set_title('Statewise Total Reported Cases')
plt.show()
print(cr)
```

```python
Indiana_Victims=dataset[dataset['State']=='Indiana']
Indiana_Victims.head()
```

```python
import squarify

p=dataset.groupby('State').sum().sort_values(by='#Total',ascending=False).head(15)
y=p['#Total']
plt.figure(figsize=(12,6))
squarify.plot(y,label=list(zip(p.index,p['#Total'])),alpha=.6,pad=False, color=plt.cm.magma(np.linspace(0,1,15)))
plt.axis('off')
plt.title('Tree Map for Top Crime', fontsize=12)
```

```python
dataset['Incident date']=pd.to_datetime(dataset['Incident Date'])
dataset['Year']=dataset['Incident Date'].dt.year
dataset['Year'].value_counts().plot.pie(figsize=(15,8))
plt.title('Crime Count each year',fontsize=15)
```

```python
dataset['Month']=dataset['Incident date'].dt.month
sns.countplot(dataset['Month'],palette='autumn')
plt.title('Crimes in each Month', fontsize=12)
```

```python
plt.figure(figsize=(30,6))
sns.boxplot(x='State',y='#Total',data=dataset)
plt.ylim(0,20)
plt.title('Crime Distribution')
```

```python
plt.figure(figsize=(30,6))
sns.scatterplot(x='State',y='#Total',data=dataset)
plt.ylim(0,30)
```
